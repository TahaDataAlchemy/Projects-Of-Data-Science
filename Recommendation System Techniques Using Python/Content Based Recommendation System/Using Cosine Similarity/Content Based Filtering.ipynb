```python
import pandas as pd
import numpy as np
import ast
```

```python
movies = pd.read_csv("tmdb_5000_movies.csv")
```

```python
credit = pd.read_csv("tmdb_5000_credits.csv")
```

```python
credit.head(1)["cast"].values
```

```python
df = movies.merge(credit ,on = "title")
```

```python
df.shape
```

```python
df
```

```python
df.columns
```

```python
# taking only important columns
df = df[["genres",'id','keywords', 'title','overview','production_companies','cast','crew']]
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df.iloc[0].genres
```

```python
def convert(obj):
    l = []
    for i in ast.literal_eval(obj):
        l.append(i["name"])
    return l 
    
```

```python
df["genres"] = df["genres"].apply(convert)
```

```python
df["keywords"] = df["keywords"].apply(convert)
```

```python
def convert3(obj):
    l = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter!=3:
            l.append(i["name"])
        else:
            break
    return l
```

```python
df["cast"] = df["cast"].apply(convert3)
```

```python
def fetch_director(obj):
    d = []
    for i in ast.literal_eval(obj):
        if i["job"] == "Director":
            d.append(i["name"])
            break
    return d
```

```python
df["crew"] = df["crew"].apply(fetch_director)
```

```python
df["overview"] = df["overview"].apply(lambda x:x.split())
```

```python
df["genres"] = df["genres"].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
df["keywords"] = df["keywords"].apply(lambda x:[i.replace(" ","") for i in x])
df["cast"] = df["cast"].apply(lambda x:[i.replace(" ","") for i in x])
df["crew"] = df["crew"].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
df["Tags"] = df["overview"]+df["genres"]+df["keywords"]+df["cast"]+df["crew"]
```

```python
new_df = df[["id","title","Tags"]]
```

```python
new_df
```

```python
new_df["Tags"] = new_df["Tags"].apply(lambda x:" ".join(x))
```

```python
new_df
```

```python
new_df["Tags"] = new_df["Tags"].apply(lambda x:x.lower())
```

```python
new_df.sample(5)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

import nltk 

from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stem(obj):
    y = []
    for i in obj.split():
        y.append(ps.stem(i))
    return " ".join(y)
        
```

```python
new_df["Tags"] = new_df["Tags"].apply(stem)
```

```python
cv = CountVectorizer(max_features = 5000,stop_words="english")
```

```python
cv
```

```python
Vectors = cv.fit_transform(new_df["Tags"]).toarray()
```

```python
Vectors[0]
```

```python
Vector5000 = cv.get_feature_names()
```

```python
Vector5000
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(Vectors)
```

```python
def recommended(movie):
    movie_index = new_df[new_df["title"] == movie].index[0]
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)),reverse=True,key = lambda x:x[1])[1:6]
    
    for i in movie_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommended('Sex and the City 2')
```
